## Function that averages each feature for each year

```python
def avg_each_year(df):
    
    sequential = df.copy()
    group = sequential.groupby("year")
    average_each_year = group.mean()
    average_each_year = average_each_year.reset_index()
    
    return average_each_year

average_year = avg_each_year(data)
avg_year = average_year.copy()
avg_year.tail()
```

## Line Plots

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(20, 8))  # chart 19.5in x 8
ax.set_title("A Few Key Features (after scaling)", 
             fontsize=14, fontweight='bold')

plt.plot(avg_year['year'],avg_year['acousticness'], color='black', label='acousticness')
plt.plot(avg_year['year'], avg_year['danceability'], color='red', label='danceability')
plt.plot(avg_year['year'], avg_year['energy'], color='blue', label='energy')
plt.plot(avg_year['year'], avg_year['instrumentalness'], color='purple', label='instrumentalness', linestyle="--")
plt.plot(avg_year['year'], avg_year['loudness'], color='c', label='loudness')
plt.plot(avg_year['year'], avg_year['valence'], color='green', label='valence')

plt.legend(loc="upper right")
plt.show()
```
